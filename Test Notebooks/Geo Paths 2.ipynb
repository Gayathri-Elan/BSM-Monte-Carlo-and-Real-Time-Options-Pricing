```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
S0 = 100 # initial stock price
r = 0.05 # risk-free rate
sigma = 0.2 # volatility
T = 1 # time to maturity
M = 252 # number of time steps
N = 1000 # number of simulations

# Define function for simulating stock prices
def simulate_stock_price(S0, r, sigma, T, M):
    dt = T/M
    price = np.zeros(M+1)
    price[0] = S0
    for i in range(1, M+1):
        price[i] = price[i-1] * np.exp((r-0.5*sigma**2)*dt + sigma*np.sqrt(dt)*np.random.normal())
    return price

# Simulate stock prices for N simulations
stock_prices = np.zeros((N, M+1))
for i in range(N):
    stock_prices[i,:] = simulate_stock_price(S0, r, sigma, T, M)

# Plot stock price paths
fig, ax = plt.subplots()
ax.plot(stock_prices.T)
ax.set_xlabel("Time")
ax.set_ylabel("Stock Price")
ax.set_title("Stock Price Simulation")
plt.show()
```
