```python
import yfinance as yf

msft = yf.Ticker("MSFT")

msft.info
```

```python
import datetime
dividend_date_timestamp = msft.info['dividendDate']
dividend_date = datetime.datetime.utcfromtimestamp(dividend_date_timestamp).strftime('%Y-%m-%d')
print("Microsoft's next dividend payment is scheduled for:", dividend_date)
```

```python
#get historical market data
start_date = "2022-12-31"
end_date = "2023-04-29"
hist = yf.download('MSFT', start=start_date, end=end_date)
print(hist)
```

```python
hist = msft.history(period="1mo")
print(hist)
```
